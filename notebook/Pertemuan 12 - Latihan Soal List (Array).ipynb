# Soal 1
Buatlah program untuk menampilkan item unik yang ada di dalam sebuah array. Ukuran array bebas ditentukan oleh user. Baris pertama merupakan ukuran array.

**Contoh 1:**  
Input
```
4
1
2
3
4
```
Output
```
1 2 3 4
```

**Contoh 2:**  
Input
```
4
2
2
3
1
```
Output
```
3 1
```
## Algoritma
1. Masukkan input jumlah item array ke dalam variabel n.
2. Buat sebuah variabel array → arr
3. Lakukan for loop untuk memasukkan input ke dalam array menggunakan fungsi append().
4. Tambahkan sebuah variabel array untuk menyimpan hasil count() tiap item → arrcount.  Fungsi count() akan menghasilkan berapa jumlah item tersebut di dalam sebuah array.
5. Lakukan for loop untuk mengisi hasil count tiap item ke dalam arrcount.  
Contoh hasil akhir variabelnya:   
```
arr = [1, 2, 3, 4]
arrcount = [1, 1, 1, 1]
arr = [2, 2, 3, 1]
arrcount = [2, 2, 1, 1]
```
5. Lakukan for loop untuk menampilkan item yang jumlah count-nya hanya 1.

```
Output contoh 1
1 2 3 4

Output contoh 2
3 1
```
Penjelasan:  
Pada contoh 1, karena semua item count-nya 1 maka semua item adalah unik dan ditampilkan semua.  
Pada contoh 2, item pertama dan kedua adalah `2` sehingga masing-masing count-nya adalah 2. Sedangkan item `3` dan `1` count-nya hanya 1. Sehingga item yang ditampilkan hanya `3` dan `1`.

Contoh array dinamis: https://github.com/munif/alpro-ti/blob/master/samples/1617/list/list-dinamis.py

# Soal 2
Buatlah sebuah program untuk melakukan rotasi array 1 dimensi sesuai dengan jumlah rotasi yang diinginkan. Baris pertama merupakan `n`, yaitu input ukuran array. Kemudian diikuti `n` baris data array. Kemudian setelah itu adalah `m`, yaitu jumlah rotasi yang diinginkan.

**Contoh 1**:  
Input
```
3
1
2
3
1
```
Output
```
2 3 1
```
**Penjelasan**  
Karena yang diinginkan adalah rotasi sebanyak 1 kali, maka tiap item akan maju sebanyak 1 posisi ke depan. Sehingga item **2** menjadi berpindah ke indeks 0, item **3** berpindah ke indeks 1, dan item 1 yang awalnya di depan berpindah ke paling belakang.

**Contoh 2**:  
Input
```
4
1
2
3
4
2
```
Output
```
3 4 1 2
```
**Penjelasan:**  
Karena mengalami rotasi sebanyak 2 kali, maka item **3** maju dua indeks ke depan, diikuti dengan item yang lain. Item-item yang sebelumnya di depan dipindah ke belakang.

Tuliskan algoritma yang digunakan kemudian buatlah programnya.
